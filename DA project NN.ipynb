```
import pandas as pd
import numpy as np
```

```
data_df = pd.read_csv('/content/cardio_train.csv',delimiter=';')
data_df.head()
```

```
data_df = data_df.drop(['id'],axis=1)
data_df.age = data_df.age//365
data_df.head()
```

```
X = data_df.iloc[:,0:11]
y = data_df.iloc[:,11]
```

```
columns = ['gender','smoke','alco','active']
Xn=pd.get_dummies(X,columns=columns)
Xn.head()
```

```
Xn.head()
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(Xn, y, test_size=0.2,random_state=42)
```

```
y_train.head()
```

```
X_train.head()
```

```
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
X_train[['age','height','weight','ap_lo','ap_hi']] = mms.fit_transform(X_train[['age','height','weight','ap_lo','ap_hi']])
X_test[['age','height','weight','ap_lo','ap_hi']] = mms.transform(X_test[['age','height','weight','ap_lo','ap_hi']])
```

```
X_train.head(10)
```

```
X_train_values = X_train.values
X_test_values = X_test.values
```

```
X_train_values[0]
```

```
import tensorflow as tf
```

```
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10,input_dim=15,activation='relu'),
    tf.keras.layers.Dense(8,activation='relu'),
    tf.keras.layers.Dense(8,activation='relu'),
    tf.keras.layers.Dense(4,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid'),   
])
```

```
model.compile(optimizer = 'adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
model.fit(X_train_values,y_train,epochs=10)
```

```
model.evaluate(X_test_values,  y_test, verbose=2)
```

```
y_pred = model.predict(X_test)
y_pred = (y_pred>0.5)
```

```
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
